Creating the Dim tables

```python
#Create gold schema if not exists
spark.sql("CREATE SCHEMA IF NOT EXISTS final_project.gold")
```

Creating Dim_Product


```python
#Storing the silver product table to a dataframe
df_silver_product = spark.read.table("final_project.silver.product")
display(df_silver_product)
```

```python
#Creating the dim_product table with spark
#Save silver table product in the gold schema as a delta table
df_silver_product.write.mode("overwrite").saveAsTable("final_project.gold.dim_product")

```

```python
#Make sure the product key is not null
spark.sql("ALTER TABLE final_project.gold.dim_product ADD CONSTRAINT product_key_not_null CHECK (product_key IS NOT NULL)")
```

```python
#Set product_key as PRIMARY KEY
spark.sql("ALTER TABLE final_project.gold.dim_product ADD CONSTRAINT product_key_pk PRIMARY KEY (product_key) RELY")
```
