**Salesperson Region:** Creating a new data frame, to cleanse the raw data, and make the necessary transformations.

```python
df_salesperson_region_clean = spark.read.csv("/Volumes/final_project/bronze/data/SalespersonRegion.csv", inferSchema=True, header=True)
display(df_salesperson_region_clean)
```

Fixing the delimiter issue (entries separated by tab) in order to create the appropriate columns.

```python
df_salesperson_region_clean = spark.read.format("csv") \
    .option("header", "true") \
    .option("sep", "\t") \
    .load("/Volumes/final_project/bronze/data/SalespersonRegion.csv")
```

```python
df_salesperson_region_clean.display()
```

```python
df_salesperson_region_clean.printSchema()
```

Changing the data types for Employee Key and Sales Territory Key because they should be numerical.

```python
from pyspark.sql.functions import col

df_salesperson_region_clean = df_salesperson_region_clean.withColumn("EmployeeKey", col("EmployeeKey").cast("int"))
df_salesperson_region_clean = df_salesperson_region_clean.withColumn("SalesTerritoryKey", col("SalesTerritoryKey").cast("int"))
```

```python
df_salesperson_region_clean.printSchema()
```

Changing the column names.

```python
df_salesperson_region_clean = df_salesperson_region_clean.toDF(
    "employee_key",
    "sales_territory_key"
)
df_salesperson_region_clean.display()
```

Checking to see if there are null values for employee key & sales territory key.

```python
df_salesperson_region_clean.filter(col("employee_key").isNull()).display()
```

```python
df_salesperson_region_clean.filter(col("sales_territory_key").isNull()).display()
```

Creating the delta table Salesperson region.

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS final_project.silver")

df_salesperson_region_clean.write.format("delta") \
  .mode("overwrite") \
  .saveAsTable("final_project.silver.salesperson_region")

display(spark.table("final_project.silver.salesperson_region"))
```
